```python

number_noise = list(range(1, 6))

jumpmean_td_l2 = [0.809, 0.646,	0.82,	0.028039753,	0.034298374]
jumpmean_td_dmd = [0.851,	0.87,	0.854,	0.846,	0.01961387]
jumpmean_td_baseline = 0.863
jumpmean_td = {'baseline': jumpmean_td_baseline, 'l2': jumpmean_td_l2, 'dmd': jumpmean_td_dmd}

jumpmean_fd_l2 = [0.635 ,0.655,	0.654, 0 , 0]
jumpmean_fd_dmd = [0.849,	0.855,	0.841,	0.839,	0.01957984 ]
jumpmean_fd_baseline = 0.837
jumpmean_fd = {'baseline': jumpmean_fd_baseline, 'l2': jumpmean_fd_l2, 'dmd': jumpmean_fd_dmd}


jumpmean_both_l2 = [0.751, 0.65,	0.758, 0 , 0]
jumpmean_both_dmd = [0.853, 0.872,0.847 ,0.841, 0 ]
jumpmean_both_baseline = 0.853
jumpmean_both = {'baseline': jumpmean_both_baseline, 'l2': jumpmean_both_l2, 'dmd': jumpmean_both_dmd}

jumpmean = {'TD': jumpmean_td, 'FD': jumpmean_fd, 'both': jumpmean_both}


fig, ax = plt.subplots(1, 3, figsize=(12, 3))
for i, value in enumerate(['TD', 'FD', 'both']):
    data = jumpmean[value]
    ax[i].plot(number_noise, data['l2'], label = f'{value}_L2')
    ax[i].plot(number_noise, data['dmd'], label=f'{value}_DMD')
    ax[i].hlines(y=data['baseline'], xmin=1, xmax=5, linestyles='--', colors='k', label='no noise')
    ax[i].legend(loc='lower left')
    ax[i].set_xlabel('number of noise channel')
    ax[i].set_ylabel('F1')
    ax[i].set_xticks(number_noise)
    ax[i].set_title(f'F1 score For {value}')
plt.show()
```

## gaussian noise 

```python

number_noise = list(range(1, 6))

td_l2 = [0.439, 0.442,0.386,0.569, 0.468]
td_dmd = [0.931, 0.398, 0.389, 0.424 ,0.409]
td_baseline = 0.95
td = {'baseline': td_baseline, 'l2': td_l2, 'dmd': td_dmd}

fd_l2 = [0.393,0.422,0.419,0.415 ,0.406]
fd_dmd = [0.918, 0.408, 0.426,0.41, 0.42]
fd_baseline = 0.936
fd = {'baseline': fd_baseline, 'l2': fd_l2, 'dmd': fd_dmd}


both_l2 = [0.43, 0.442, 0.417, 0.52, 0.452]
both_dmd = [0.953, 0.383, 0.398, 0.433,	0.416]
both_baseline = 0.969
both = {'baseline': both_baseline, 'l2': both_l2, 'dmd': both_dmd}

all_data = {'TD': td, 'FD': fd, 'both': both}


fig, ax = plt.subplots(1, 3, figsize=(12, 3))
for i, value in enumerate(['TD', 'FD', 'both']):
    data = all_data[value]
    ax[i].plot(number_noise, data['l2'], label = f'{value}_L2')
    ax[i].plot(number_noise, data['dmd'], label=f'{value}_DMD')
    ax[i].hlines(y=data['baseline'], xmin=1, xmax=5, linestyles='--', colors='k', label='no noise')
    ax[i].legend(loc='center right')
    ax[i].set_xlabel('number of noise channel')
    ax[i].set_ylabel('F1')
    ax[i].set_xticks(number_noise)
    ax[i].set_title(f'F1 score For {value}')
plt.show()
```

```python
'''
With Gaussian noise, the DMD remain nearly the same performance when comparing with Gaussian mixtures with no noise. However, the performance drop significantly when adding more than 1 noise channel. The L2-norm doesn't work well with Gausian noise. 

DMD can deal better with AR noise but not good with Gaussian noise. 
'''
```
